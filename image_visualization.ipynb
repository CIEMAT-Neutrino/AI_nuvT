```python
import numpy as np
import matplotlib.pyplot as plt

event_index = np.array([5760])  # Event index array
num_columns = 4 # Fixed number of columns
num_rows = 2  # Number of rows

labels = [
    "PE1: Visible/Volume -", "PE2: Visible/Volume +", 
    "PE3: Ultraviolet/Volume -", "PE4: Ultraviolet/Volume +", 
    "T1: Visible/Volume -", "T2: Visible/Volume +", 
    "T3: Ultraviolet/Volume -", "T4: Ultraviolet/Volume +"
]

# Create the figure and axes
fig, axs = plt.subplots(num_rows, num_columns, figsize=(25, 5))

# Flatten the axes for easier indexing
axs = axs.flatten()

# Loop through the total number of labels
for idx in range(len(labels)):
    # Access the image using event_index and the correct axis
    axs[idx].imshow(np.squeeze(image_min[event_index[0], :, :, idx]), cmap='Blues')
    axs[idx].set_title(labels[idx],fontsize=24)  # Set the title for each subplot

# Remove ticks
plt.setp(axs, xticks=[], yticks=[])

plt.show()
```

**Comparison of images equivalence, image_gen is the new function**

```python
# Get the shape of the dimensions to iterate over
num_events, num_rows, num_cols, _ = image_min.shape

# Initialize a list to store the results
differences = []

# Iterate over all possible values of i, a, and b
for i in range(num_events):
    for a in range(num_rows):
        for b in range(num_cols):
            # Compare the values at the current indices
            if image_min[i][a][b][3] != image_min_new[i][a][b][5]:
                differences.append((i, a, b, image_min[i][a][b][1], image_min_new[i][a][b][1]))

# Output the differences
if differences:
    for diff in differences:
        print(f"Difference found at (i={diff[0]}, a={diff[1]}, b={diff[2]}): "
              f"image_min = {diff[3]}, image_min_new = {diff[4]}")
else:
    print("All values are equal for the specified indices.")
```
