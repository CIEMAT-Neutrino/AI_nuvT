```python
# Get the shape of the dimensions to iterate over
num_events, num_rows, num_cols, _ = image_min.shape

# Initialize a list to store the results
differences = []

# Iterate over all possible values of i, a, and b
for i in range(num_events):
    for a in range(num_rows):
        for b in range(num_cols):
            # Compare the values at the current indices
            if image_min[i][a][b][3] != image_min_new[i][a][b][5]:
                differences.append((i, a, b, image_min[i][a][b][1], image_min_new[i][a][b][1]))

# Output the differences
if differences:
    for diff in differences:
        print(f"Difference found at (i={diff[0]}, a={diff[1]}, b={diff[2]}): "
              f"image_min = {diff[3]}, image_min_new = {diff[4]}")
else:
    print("All values are equal for the specified indices.")
```
