```python
event_index = np.array([523])

num_columns = 1  # Fixed number of columns
num_rows = 8  # Number of rows

labels = [
    "C1: Visible/Volumen -", "C2: Visible/Volumen +", 
    "C3: Ultravioleta/Volumen -", "C4: Ultravioleta/Volumen +", 
    "T1: Visible/Volumen -", "T2: Visible/Volumen +", 
    "T3: Ultravioleta/Volumen -", "T4: Ultravioleta/Volumen +"
]

fig, axs = plt.subplots(num_rows, num_columns,figsize=(30,80))

for i in range(num_rows):
    axs[i].imshow(np.squeeze(image_max[event_index, :, :, i-1]), cmap='BuPu')  # Use squeeze to fix shape
    axs[i].set_title(labels[i])  # Set the title for each subplot

plt.setp(axs, xticks=[], yticks=[])
plt.show()
```
